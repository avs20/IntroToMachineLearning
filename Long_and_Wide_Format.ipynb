```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
%matplotlib inline
```

```
data = pd.read_csv('suicide_total_deaths.csv')
data.head()
```

```
data.plot()
```

```
data_ind = data.loc[data.country == 'India', :]
data_ind.head()
```

```
pd.DataFrame(data_ind.T).reset_index()
```

```
data_t = pd.DataFrame(data_ind.T).reset_index()
data_t.rename(columns = {'index':'year',75:'deaths'}, inplace=True)
data_t.drop(index = 0, inplace=True)
data_t.plot(x = 'year', y = 'deaths')
```

```
pd.melt(data_ind,var_name='year',id_vars='country')
```

```
pd.melt(data_ind,var_name='year',id_vars='country').loc[1:,:].plot()
```
